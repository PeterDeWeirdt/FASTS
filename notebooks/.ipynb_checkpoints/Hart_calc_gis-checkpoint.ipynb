```python
import pandas as pd
import gnt
import gpplot
import matplotlib.pyplot as plt
import stringdb
import seaborn as sns
import networkx as nx
import numpy as np
```

```python
gpplot.set_aesthetics(context = 'paper')
```

```python
lfcs = pd.read_csv('../data/interim/FASN_avg_lfcs.csv')
qgis = pd.read_csv('../data/external/moffat_hap1_qgi.csv', header=1)
library_design = pd.read_csv('../data/raw/Cas12_FASN_GI_12k_library.csv')
```

```python
control_genes = ['CXorf66', 'GPX6', 'BMP15', 'RXFP3', 'PRSS37', 'TSSK1B', 'SLC36A3', 'C3orf30',
                 'NDST4', 'RNF17', 'OPN5', 'RAX', 'GSX2', 'GLRA2', 'PPP3R2']
```

```python
guide_residuals, model_info = gnt.get_guide_residuals(lfcs, control_genes)
```

```python
gene_residuals = gnt.get_gene_residuals(guide_residuals, 'residual_z')
```

```python
gene_residuals.sort_values('pair_z_score').head(50)
```

```python
utils.label_anchor_target_gene(guide_residuals, 'SREBF1', 'SREBF2')
gpplot.savefig('../figures/SREBF1_SREBF2_sactter.png')
```

```python
spread_gene_residuals = (gene_residuals.pivot_table(index=['gene_a', 'gene_b', 'guide_pairs'], 
                                                    values='pair_z_score', 
                                                    columns='condition')
                         .reset_index())
spread_gene_residuals
```

```python
spread_gene_residuals['gene_pair'] = spread_gene_residuals['gene_a'] + '_' + spread_gene_residuals['gene_b']
```

```python
fig, ax = plt.subplots(figsize = (2.2,2.2))
ax = gpplot.point_densityplot(spread_gene_residuals, "MOLM13_14", "MOLM13_21", ax = ax)
gpplot.add_correlation(spread_gene_residuals, "MOLM13_14", "MOLM13_21")
plt.title('Genetic interaction z-score')
gpplot.savefig('../figures/MOLM13_correlation.svg', dpi=300)
```

```python
top_n = 5
top_MOLM13 = spread_gene_residuals[spread_gene_residuals.MOLM13_21.abs().rank(ascending=False) <= top_n]
top_MOLM13
```

```python
fig, ax = plt.subplots(figsize = (2.2,2.2))
ax = gpplot.point_densityplot(spread_gene_residuals, "NOMO-1_14", "NOMO-1_21", ax = ax)
gpplot.add_correlation(spread_gene_residuals, "NOMO-1_14", "NOMO-1_21")
plt.title('Genetic interaction z-score')
gpplot.savefig('../figures/NOMO_correlation.svg', dpi=300)
```

```python
top_NOMO1 = spread_gene_residuals[spread_gene_residuals['NOMO-1_21'].abs().rank(ascending=False) <= top_n]
top_NOMO1
```

```python
edge_cutoff = 3
fig, ax = plt.subplots(figsize = (3.5,3.5))
ax = gpplot.point_densityplot(spread_gene_residuals, "MOLM13_21", "NOMO-1_21", ax = ax)
gpplot.add_correlation(spread_gene_residuals, "MOLM13_21", "NOMO-1_21")
gpplot.label_points(spread_gene_residuals, "MOLM13_21", "NOMO-1_21", 
                    set(top_NOMO1.gene_pair.to_list() + top_MOLM13.gene_pair.to_list()), 'gene_pair')
plt.axvline(x=-edge_cutoff, linestyle='dashed', color='black')
plt.axvline(x=edge_cutoff, linestyle='dashed', color='black')
plt.axhline(y=-edge_cutoff, linestyle='dashed', color='black')
plt.axhline(y=edge_cutoff, linestyle='dashed', color='black')
plt.title('Genetic interaction z-score')
gpplot.savefig('../figures/NOMO_MOLM13_correlation.png', dpi=300)
```

## Clustering

```python
library_genes = library_design[['GENE_1', 'GENE_2']].drop_duplicates()
remapped_genes = gnt.score.order_cols(library_genes, [0, 1], 'gene')
remapped_gene_residuals = gene_residuals.merge(remapped_genes, how='inner', on=['gene_a', 'gene_b'])
```

```python
condition = 'MOLM13_21'
residual_df = remapped_gene_residuals

residual_df = residual_df.copy()
condition_residuals = residual_df[residual_df.condition == condition]
spread_condition = condition_residuals.pivot_table(index='GENE_1', columns='GENE_2', )
```

## Network analysis

```python
molm13_edgelist = gene_residuals[(gene_residuals.z_score_residual_z.abs() > edge_cutoff) & 
                                 (gene_residuals.condition == 'MOLM13_21')].copy()
molm13_network = utils.generate_graph(molm13_edgelist, 'z_score_residual_z')
```

```python
fig, ax = plt.subplots(figsize = (7, 7))
g = utils.draw_graph(molm13_network, ax,  'color', 'abs_z')
plt.title('MOLM13 |z-score| > ' + str(edge_cutoff))
gpplot.savefig('../figures/MOLM13_network.png')
```

```python
nomo1_edgelist = gene_residuals[(gene_residuals.z_score_residual_z.abs() > edge_cutoff) & 
                                 (gene_residuals.condition == 'NOMO-1_21')].copy()
nomo1_network = utils.generate_graph(nomo1_edgelist, 'z_score_residual_z')
fig, ax = plt.subplots(figsize = (7, 7))
g = utils.draw_graph(nomo1_network, ax, 'color', 'abs_z')
plt.title('NOMO-1 |z-score| > '+ str(edge_cutoff))
gpplot.savefig('../figures/NOMO1_network.png')
```

## Comparison with HAP1 single cell clones

```python
melted_qgis = (qgis
                  .rename({'FASN_merge': 'FASN'}, axis = 1)
                  .drop(['qGI_FASN_a', 'qGI_FASN_b', 'qGI_FASN_c'], axis = 1)
                  .melt(id_vars = 'gene', var_name='gene2', value_name='qgi'))
melted_qgis = (gnt.score.order_genes(melted_qgis, gene_cols=[0, 1])
                 .drop(['gene', 'gene2'], axis=1))
melted_qgis['gene_pair'] = melted_qgis['gene_a'] + '_' + melted_qgis['gene_b'] 
```

```python
melted_qgis
```

```python
joined_qgis = melted_qgis.merge(spread_gene_residuals, how='inner', on = ['gene_a', 'gene_b', 'gene_pair'])
```

```python
top_qgis = joined_qgis[joined_qgis.qgi.abs().rank(ascending=False) <= top_n]
top_qgis
```

```python
fig, ax = plt.subplots(figsize = (9,9))
ax = gpplot.point_densityplot(joined_qgis, "MOLM13_21", "qgi", ax = ax)
ax = gpplot.add_correlation(joined_qgis, "MOLM13_21", "qgi")
ax = gpplot.label_points(joined_qgis, "MOLM13_21", "qgi",  
                         set(top_qgis.gene_pair.to_list() + top_MOLM13.gene_pair.to_list()), 
                         "gene_pair")
plt.ylabel('Moffat HAP1 (qGI)')
plt.xlabel('MOLM13_21 (z-score)')
plt.title('HAP1 KO vs MOLM13 combo')
gpplot.savefig('../figures/MOLM13_qgi_correlation.png')
```

```python
fig, ax = plt.subplots(figsize = (9,9))
ax = gpplot.point_densityplot(joined_qgis, "NOMO-1_21", "qgi", ax = ax)
ax = gpplot.add_correlation(joined_qgis, "NOMO-1_21", "qgi")
ax = gpplot.label_points(joined_qgis, "NOMO-1_21", "qgi",  
                         set(top_qgis.gene_pair.to_list() + top_NOMO1.gene_pair.to_list()), 
                         "gene_pair")
plt.ylabel('Moffat HAP1 (qGI)')
plt.xlabel('NOMO1_21 (z-score)')
plt.title('HAP1 KO vs NOMO1 combo')
gpplot.savefig('../figures/NOMO1_qgi_correlation.png')
```
